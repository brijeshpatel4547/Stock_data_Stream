```
df = spark.read.format("com.databricks.spark.avro").load('/mnt/stockdata/stock-market-analytics/stock-stream_0_2024_03_29/*.avro')
```

```


# Dropping unnecessary feilds
cols = ['SequenceNumber','Offset','EnqueuedTimeUtc','SystemProperties','Properties']
df = df.drop(*cols)
df.collect()
```

```
# decoding the byte array records into ascii values and adding them in a list

lst = []
for i in df.collect():
    lst.append(i['Body'].decode('ascii'))
print(lst)
```

```
#Creating the RDD out of the list 
rdd1 = sc.parallelize(lst)
rdd2 = rdd1.map(lambda x : x.split(","))
```

```
#RDD to Dataframe
newcols = ['Index','Date','Open','High','Low','Close','Adj Close','Volumne','CloseUSD']
df1 = rdd2.toDF(newcols)
df1.count()
```

```
df1 = df1.withColumnRenamed('Volumne','Volume')
```

```
df1.show(5)
```

```
df1.printSchema()
```

```
# Updating the schema of the dataframe.
df1 = df1.withColumn("Date",col("Date").cast("Date"))
df1 = df1.withColumn("Open",col("Open").cast("float"))
df1 = df1.withColumn("High",col("High").cast("float"))
df1 = df1.withColumn("Low",col("Low").cast("float"))
df1 = df1.withColumn("Close",col("Close").cast("float"))
df1 = df1.withColumn("Adj Close",col("Adj Close").cast("float"))
df1 = df1.withColumn("CloseUSD",col("CloseUSD").cast("float"))
df1 = df1.withColumn("Volume",col("Volume").cast("integer"))
```

```
df1.repartition(1).write.mode("overwrite").option("header",'true').csv("/mnt/stockdata/transformed/index")
```
